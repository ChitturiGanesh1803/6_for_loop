```python
# find factors of 10(LCM)
n = int(input("Enter a number"))
for x in range(1,n):
    if n%x == 0:
        print(x)
```

```python
#prime no or not prime
n = int(input("enter a number"))
count = 0
for i in range(1,n+1):
    if n%i == 0:
        count += 1
if count == 2:
    print("Prime_no")
else:
    print("not prime_no")
```
